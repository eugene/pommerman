```python
import pommerman
from pommerman import agents
import numpy as np
from copy import deepcopy
from matplotlib import pyplot as plt
from tqdm import tqdm_notebook as tqdm
from matplotlib import pyplot as plt
from collections import Counter
```

## Plots

```python
first = [129680.,  40515.,  49923.,  42435.,  49994.,  33249.]
f_sum = np.sum(first)
```

```python
second = [299927., 323366., 503512., 337408., 372636.,  98092.]
s_sum = np.sum(second)
```

```python
third = [364500., 518445., 572591., 521393., 587000., 113351.]
t_sum = np.sum(third)
```

```python
fourth = [313905., 427717., 491914., 426943., 488402.,  94898.]
fth_sum = np.sum(fourth)
```

```python
N = len(first) # the x locations for the groups

ind = np.arange(N)
width = 0.25       # the width of the bars

fig = plt.figure(figsize=(18,8))
plt.bar(ind, first/f_sum, width, color='b')

plt.bar(ind + width, second/s_sum, width, color='orange')

plt.bar(ind + width*2, third/t_sum, width, color='g')

plt.xticks(np.arange(6), ('Stop','Up','Down','Left','Right','Bomb'))

plt.title('Each game fraction', fontsize = 30)
plt.tick_params(axis = 'both', which = 'major', labelsize = 24)

plt.legend(['Simple vs three random', 'Simple vs one simple and two random', 'Simple vs three simple'], loc=0)

plt.savefig('fig1.png', dpi=fig.dpi)
```

```python
together = [first, second, third]
```

```python
sumindex = [sum(elts) for elts in zip(*together)]
```

```python
N = len(first) # the x locations for the groups

ind = np.arange(N)
width = 0.25       # the width of the bars

fig = plt.figure(figsize=(18,8))

plt.bar(ind, [x/y for x, y in zip(first, sumindex)], width, color='b')

plt.bar(ind + width, [x/y for x, y in zip(second, sumindex)], width, color='orange')

plt.bar(ind + width*2, [x/y for x, y in zip(third, sumindex)], width, color='g')

plt.xticks(np.arange(6), ('Stop','Up','Down','Left','Right','Bomb'))
#plt.ylim(0,1.15)

plt.title('Each action fraction', fontsize = 30)
plt.tick_params(axis = 'both', which = 'major', labelsize = 24)

plt.legend(['Simple vs three random', 'Simple vs one simple and two random', 'Simple vs three simple'], loc=0)

plt.savefig('fig2.png', dpi=fig.dpi)
```
